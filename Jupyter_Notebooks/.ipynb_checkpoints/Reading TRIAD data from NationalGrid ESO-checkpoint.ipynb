```python
import os
triad_files = os.listdir('../Triad/')
triad_files
```

I have 13 years of Triad data, however they are not in csv format(nor in excel) so I will need to find a way to read this data. Luckily there is a library called tabula which helps to read tables from pdf files.

```python
from tabula import read_pdf #for reading pdf files
import pandas as pd
```

```python
triad_dfs = {}
file_dir = "../Triad/"
for i in triad_files:

    year = i.split(" ")[0]
    year = ''.join(filter(str.isdigit, year))
    year = year[0:4]
    print("Getting the table for year",year)
    list_df = read_pdf(file_dir+i,pages = "all",multiple_tables=True)
    #I know there is some problem with 2017, I do not know now exactly, but for now this will work
    if year == "2017":
        df = list_df[1]
    else:
        df = list_df[0]
    df['year'] = year
    triad_dfs[year] = df
    print("Finished for year",year)
```
